### iv. Encoding Categorical Data

```python
# Change the all categorical values into numeric values as ML supports only numericals
```

```python
# Here the hypothesis is 'Make','Model' won't effect the 'CO2 Emissions' but specifications does

X = df[['Vehicle Class','Transmission','Fuel Type','Engine Size(L)','Cylinders','Fuel Consumption Comb (L/100 km)']]
Y = df[['CO2 Emissions(g/km)']]

X.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
col = ['Vehicle Class','Transmission','Fuel Type']
oe = OrdinalEncoder()
X[col] = oe.fit_transform(X[col])
```

```python
X
```

### v. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
cols = ['Engine Size(L)','Cylinders','Fuel Consumption Comb (L/100 km)']
X[cols]=sc.fit_transform(X[cols])
print(X.shape)
X
```

```python
print(Y.shape)
Y
```

### vi. Splitting the dataset into test set and training set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.45)
```

```python
plt.figure(figsize=(12,12),dpi=200)
order = df.groupby(['Make']).mean().sort_values('CO2 Emissions(g/km)',ascending=False).index
sns.barplot(df['CO2 Emissions(g/km)'],df['Make'], order = order)
```

```python
plt.figure(dpi=200)
order = df.groupby(['Fuel Type']).mean().sort_values('CO2 Emissions(g/km)',ascending=False).index
sns.barplot(df['Fuel Type'],df['CO2 Emissions(g/km)'], order = order)
```

```python
plt.figure(figsize=(10,20),dpi=200)
order = df.groupby(['Make']).mean().sort_values('CO2 Emissions(g/km)',ascending=False).index
sns.barplot(df['CO2 Emissions(g/km)'],df['Make'],hue = df['Fuel Type'], order = order)
```

# Observation : 
   ## The vehicles Made by  ['BUGATTI', 'LAMBORGHINI', 'SRT', 'ROLLS-ROYCE', 'BENTLEY','ASTON MARTIN',] with fuel type ['Z'] are producing the higher Emissions

```python
plt.figure(figsize=(8,10),dpi=200)
Order = df.groupby(['Vehicle Class']).mean().sort_values('CO2 Emissions(g/km)',ascending = False).index
sns.barplot(df['CO2 Emissions(g/km)'],df['Vehicle Class'],hue = df['Fuel Type'],order = Order)
```

# Observation :
   ## The Vehicle Class 'VAN - PASSENGER', 'VAN - CARGO' with fuel type 'X','E' produces high Emissions

```python
plt.figure(figsize=(12,6),dpi=200)
orde = df.groupby('Transmission').mean().sort_values('CO2 Emissions(g/km)',ascending = False).index
chart = sns.barplot(df['Transmission'],df['CO2 Emissions(g/km)'], order = orde)
chart.axhline(200)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

# Observation : 
   ## The transmissions 'M5', 'AV7','AV6', 'AM6', 'AV10', 'AV', 'AM5' are with minimum affordable emissions

```python
plt.figure(figsize=(15,7))
sns.barplot(df['Cylinders'],df['CO2 Emissions(g/km)'],hue = df['Fuel Type'])
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.lineplot(y=df['CO2 Emissions(g/km)'],x=df['Engine Size(L)'])
```

# Observation : 
   ## As the no.of cylinders and Engine size(L) increases, the emissions will increases and high no.of cylinders uses mostly 'Z' type fuel

```python
plt.figure(figsize=(12,6),dpi=200)
sns.barplot(df['Fuel Type'],df['CO2 Emissions(g/km)'],hue=df['Cylinders'])
```

# Observation :

```python
plt.figure(figsize=(12,6),dpi=200)
chart = sns.lineplot(df['Engine Size(L)'],df['CO2 Emissions(g/km)'],label = 'Engine_Size')
chart = sns.lineplot(df['Cylinders'],df['CO2 Emissions(g/km)'], label = 'No.of_Cylinders')
plt.xlabel(' ')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=200)
chart = sns.lineplot(df['Fuel Consumption City (L/100 km)'],df['CO2 Emissions(g/km)'],label = 'City')
chart = sns.lineplot(df['Fuel Consumption Hwy (L/100 km)'],df['CO2 Emissions(g/km)'], label = 'Hwy')
chart = sns.lineplot(df['Fuel Consumption Comb (L/100 km)'],df['CO2 Emissions(g/km)'], label = 'Comb')
plt.xlabel('Fuel Consumption (L/100 km)')
plt.show()
```
