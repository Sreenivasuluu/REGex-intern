## (a) Preprocessing the data

### i. Import the required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

### iii. Handling duplicated and Missing data

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)       # dropping the duplicated rows
```

```python
df
```

### Check for null values

```python
df.isnull().sum()
```

```python
# No null values in the dataset
```

### iv. Encoding Categorical Data

```python
# Change the all categorical values into numeric values as ML supports only numericals
```

```python
from sklearn.preprocessing import LabelEncoder
le_data =df.apply(LabelEncoder().fit_transform)
```

```python
X = le_data.iloc[:,:-1].values
Y = le_data.iloc[:,-1].values
print(X.shape)
X
```

```python
print(Y.shape)
Y
```

### v. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X[:,1:]=sc.fit_transform(X[:,1:])
print(X.shape)
X
```

```python
Y = sc.fit_transform(Y.reshape(-1,1))
print(Y.shape)
Y
```

### vi. Splitting the dataset into test set and training set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```
